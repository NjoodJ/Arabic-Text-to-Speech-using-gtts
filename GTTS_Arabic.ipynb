```python
from gtts import gTTS
import IPython.display as ipd
```

```python
Arabic_text= "يعمل شات جي بي تي على فهم ما تكتبه بدقة أيا كانت اللغة التي تستخدمها من خلال تقنية الذكاء الاصطناعي. يتعامل شات جي بي تي بطريقة تفاعلية مع المستخدم وكأن شخصا موجودا على الناحية الأخرى من المحادثة.تعتمد محركات البحث العادية على النص المكتوب فقط، لكن شات جي بي تي يقوم بتحليل الجملة التي نكتبها، ويعمل على فهم المراد منها، ويحضر الإجابات التي تحتاجها بشكل دقييق. يتميز شات جي بي تي بقدرته على تقديم إجابات شاملة بدلا من روابط لمقالات. "
print(Arabic_text)
```

```python
tts = gTTS(Arabic_text, lang='ar')
tts.save('test.mp3')
```

```python
audio_path="test.mp3"
ipd.Audio(audio_path, autoplay=True)
```
